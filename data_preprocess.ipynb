```python
def get_first_ten_names(row):
    row_list = ast.literal_eval(row)
    first_ten_unique_name = set([i['name'] for i in row_list][:10])
    first_ten_names = ", " .join(f"{name.replace(' ', '')}" for name in first_ten_unique_name)
    return first_ten_names
```

```python
movies_df['cast'] = movies_df['cast'].apply(get_first_ten_names)
movies_df['crew'] = movies_df['crew'].apply(get_first_ten_names)
movies_df['genres'] = movies_df['genres'].apply(get_first_ten_names)
movies_df['keywords'] = movies_df['keywords'].apply(get_first_ten_names)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(get_first_ten_names)
movies_df['production_companies'] = movies_df['production_companies'].apply(get_first_ten_names)
movies_df['production_countries'] = movies_df['production_countries'].apply(get_first_ten_names)
```
